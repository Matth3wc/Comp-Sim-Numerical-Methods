## So what's the downside?

* Python's execution speed may not always be as fast as that of fully compiled and lower-level languages such as _C_ and _C++_
    * Python is in fact compiled to _byte code_ and then interpreted within a virtual machine. This provides the characteristic portability as the virtual machine is platform independent     
* For some tasks it may be important to get "closer to the silicon" by using lower-level languages that are more directly mapped to the underlying hardware architecture.
    * However, numerically intensive parts of Python code can be split off into compiled extensions written in low-level programming languages such as _C_.

#### Defining a function

Functions are defined by the _def_ statememt

```python
def func(x):
    return 2*x**2
```

and evaluated for a particular value

```python
func(2)
```

This function can then be used directly  (or saved as a library file and imported)

```python
x = [ii for ii in range(10)]
y = [func(ii) for ii in x]
```

and then plotted (we have imported _matlplotlib_ in a previous cell)

```python
plt.plot(x,y)
```

---
#### lambda functions

The _def_ statement creates a function object and gives it a name. It is Python's most basic tool for reducing code redundancy.

The _lambda_ statement also creates an object but returns it as a result. This allows _in-line_ function definitions in places where _def_  statements do not work syntactically, such as passing a function as an argument

```python
f = lambda x,y: 2*x*y**2
print('f(2,3) = ', f(2,3))
```

#### File import

Data files can be imported using the _open_ and _close_ commands

The file data.dat contains the following text (string) dat

The file can be imported using the _open_, _close_ and _readlines_ commands

```python
f_in = open('data.dat', 'r')
data = f_in.readlines()
f_in.close()
```

The imported string data can then be parsed using the _split_ command

```python
x, y = [], []

for ind in range(len(data)):
    # Split each line into individual numbers (but still represented as strings)
    this_line = data[ind].split()
    # Convert the strings into floating point numbers and append them to the corresponding list
    x.append(float(this_line[0]))
    y.append(float(this_line[1]))
    
```

and then plotted (we have imported _matplotlib_ in previous cell)

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y = x**2')
plt.title('The square of the first 10 integers')
```
